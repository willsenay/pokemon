```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
stats = 'pokemon_stats.csv'
df = pd.read_csv(stats)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df['#']<152]
df.tail()
```

```python
for name in df['Name']:
    if fuzz.partial_ratio(name, 'Mega') == 100:
        df = df[df['Name']!=name]
```

```python
df.head(10)
```

```python
df.shape
```

```python
df = df.rename(columns={"Sp. Atk": "Sp_Atk", "Sp. Def": "Sp_Def", "#": "ID", "Type 1": "Type_1", "Type 2": "Type_2"})
```

```python
df.head()
```

```python
df1 = df.where((pd.notnull(df)), "None")
```

```python
df1.head()
```

```python
df1.to_csv('pokemon_app/gen1.csv', index=False)
```
