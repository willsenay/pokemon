```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
stats = 'pokemon_stats/pokemon_stats.csv'
df = pd.read_csv(stats)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df['#']<722]
df.tail()
```

```python
df['#'] = df['#'].astype(str)
```

```python
df = df.set_index('Name')
for name in df.index:
    if (fuzz.partial_ratio(name, 'Mega') == 100) & (name != 'Meganium'):
        df.loc[name,'#'] = df.loc[name,'#'] + '-mega'
```
